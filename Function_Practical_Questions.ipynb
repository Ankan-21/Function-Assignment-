1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list

Ans- We can create a Python function to sum all even numbers in a list using either a loop or a combination of the filter() and sum() functions. Below is an example function that demonstrates both approaches.

Approach 1: Using a Loop

def sum_even_numbers(numbers):  
    total = 0  
    for number in numbers:  
        if number % 2 == 0:  
            total += number  
    return total  

# Example usage  
numbers = [1, 2, 3, 4, 5, 6]  
result = sum_even_numbers(numbers)  
print(result)  # Outputs: 12 (2 + 4 + 6)  

Approach 2: Using filter() and sum()

def sum_even_numbers(numbers):  
    even_numbers = filter(lambda x: x % 2 == 0, numbers)  
    return sum(even_numbers)  

# Example usage  
numbers = [1, 2, 3, 4, 5, 6]  
result = sum_even_numbers(numbers)  
print(result)  # Outputs: 12 (2 + 4 + 6)  
Explanation
Using a Loop:

The function initializes a variable total to zero.
It iterates through each number in the input list.
If the number is even (i.e., divisible by 2), it adds that number to total.
Finally, it returns the total sum of even numbers.
Using filter() and sum():

The filter() function is used to create an iterator of even numbers by applying a lambda function that checks if each number is even.
The sum() function is then called on the filtered iterator to calculate the total sum of the even numbers.
Both approaches will work correctly to return the sum of even numbers in the input list. We can choose the one that we find more readable or suitable for our needs.


2.Create a Python function that accepts a string and returns the reverse of that string.

Ans- We can create a Python function to reverse a string in several ways. Below are a few different approaches: using slicing, using a loop, and using the reversed() function.

Approach 1: Using Slicing

def reverse_string(s):  
    return s[::-1]  

# Example usage  
input_string = "Hello, World!"  
result = reverse_string(input_string)  
print(result)  # Outputs: !dlroW ,olleH  

Approach 2: Using a Loop

def reverse_string(s):  
    reversed_str = ""  
    for char in s:  
        reversed_str = char + reversed_str  # Prepend each character  
    return reversed_str  

# Example usage  
input_string = "Hello, World!"  
result = reverse_string(input_string)  
print(result)  # Outputs: !dlroW ,olleH  

Approach 3: Using the reversed() Function

def reverse_string(s):  
    return ''.join(reversed(s))  

# Example usage  
input_string = "Hello, World!"  
result = reverse_string(input_string)  
print(result)  # Outputs: !dlroW ,olleH  

Explanation of Each Approach

Using Slicing:

The syntax s[::-1] creates a new string by taking the input string s and slicing it with a step of -1, effectively reversing it.
Using a Loop:

The function initializes an empty string reversed_str.
It iterates through each character in the input string and prepends it to reversed_str.
Finally, it returns the reversed string.
Using reversed():

The reversed() function returns an iterator that accesses the given string in the reverse order.
The join() method is then used to concatenate the characters back into a single string.



3.Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

Ans- We can create a Python function that takes a list of integers and returns a new list containing the squares of each number using various methods. Here are a few different approaches: using a loop, using list comprehension, and using the map() function.

Approach 1: Using a Loop

def square_numbers(numbers):  
    squares = []  
    for number in numbers:  
        squares.append(number ** 2)  # Append the square of the number  
    return squares  

# Example usage  
input_numbers = [1, 2, 3, 4, 5]  
result = square_numbers(input_numbers)  
print(result)  # Outputs: [1, 4, 9, 16, 25]  

Approach 2: Using List Comprehension

def square_numbers(numbers):  
    return [number ** 2 for number in numbers]  

# Example usage  
input_numbers = [1, 2, 3, 4, 5]  
result = square_numbers(input_numbers)  
print(result)  # Outputs: [1, 4, 9, 16, 25]

Approach 3: Using the map() Function

def square_numbers(numbers):  
    return list(map(lambda x: x ** 2, numbers))  

# Example usage  
input_numbers = [1, 2, 3, 4, 5]  
result = square_numbers(input_numbers)  
print(result)  # Outputs: [1, 4, 9, 16, 25]  
Explanation of Each Approach
Using a Loop:

The function initializes an empty list squares.
It iterates through each number in the input list, calculates its square, and appends it to the squares list.
Finally, it returns the list of squares.
Using List Comprehension:

This is a concise way to create a new list by applying an expression (here, squaring the number) to each item in the input list.
It generates the list in a single line of code.

Using map():

The map() function applies a given function (in this case, a lambda function that squares the number) to all items in the input list.
The result is converted back to a list using list().





4.Write a Python function that checks if a given number is prime or not from 1 to 200.

Ans- We can create a Python function to check if a given number is prime. A prime number is a number greater than 1 that has no positive divisors other than 1 and itself. Below is a function that checks if a number is prime, specifically for numbers in the range from 1 to 200.

Prime Checking Function

def is_prime(n):  
    if n <= 1:  
        return False  # 0 and 1 are not prime numbers  
    for i in range(2, int(n**0.5) + 1):  
        if n % i == 0:  
            return False  # Found a divisor, hence not prime  
    return True  # No divisors found, hence prime  

# Example usage  
for number in range(1, 201):  
    if is_prime(number):  
        print(f"{number} is prime")  
Explanation of the Code
Check for Base Cases:

The function first checks if n is less than or equal to 1. If so, it returns False since 0 and 1 are not prime numbers.
Checking Divisibility:

The function then uses a loop to check for factors of n.
It only checks for divisors up to the square root of n (using int(n**0.5) + 1). This is an efficient way to determine if a number is prime because if n is divisible by any number greater than its square root, it must also be divisible by a number less than its square root.
Return True or False:

If it finds any divisors, the function returns False. If the loop completes without finding any, it returns True, indicating that n is prime.
Example Usage
The example usage loops through numbers from 1 to 200, checks if each number is prime, and prints the result. we can run this code in our Python environment to see which numbers between 1 and 200 are prime.

Output Sample
When we run the example usage, we'll see output similar to this:

5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

Ans- We can create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms by implementing the iterator protocol. Here’s how we can do it:

Fibonacci Iterator Class

class FibonacciIterator:  
    def __init__(self, terms):  
        self.terms = terms  # Number of terms to generate  
        self.count = 0      # Counter to track the number of terms generated  
        self.a, self.b = 0, 1  # Starting values for Fibonacci sequence  

    def __iter__(self):  
        return self  # Return the iterator object itself  

    def __next__(self):  
        if self.count < self.terms:  
            result = self.a  
            self.a, self.b = self.b, self.a + self.b  # Update values for the next term  
            self.count += 1  # Increment the count  
            return result  
        else:  
            raise StopIteration  # End the iteration when the specified terms are reached  

# Example usage  
terms = 10  # Specify the number of terms you want  
fibonacci = FibonacciIterator(terms)  

for number in fibonacci:  
    print(number)  
Explanation of the Code
Class Definition:

FibonacciIterator is defined as an iterator class.
Initialization (__init__ method):

The __init__ method takes one argument, terms, which specifies how many terms of the Fibonacci sequence to generate.
It initializes a counter count to track how many terms have been generated and sets the starting values a and b to represent the first two Fibonacci numbers (0 and 1).
Iterator Method (__iter__):

The __iter__ method returns the iterator object itself. This is required to make the class an iterator.
Next Method (__next__):

The __next__ method generates the next Fibonacci number.
If the count of generated terms is less than terms, it returns the current Fibonacci number (result), updates a and b for the next Fibonacci number, increments the count, and then returns the result.
If the number of generated terms reaches the specified limit, it raises a StopIteration exception to end the iteration.
Example Usage
The example usage creates an instance of FibonacciIterator for a specified number of terms (10 in this case). It then iterates through the generated numbers and prints them.

6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

Ans- We can create a generator function in Python that yields the powers of 2 up to a specified exponent. A generator function uses the yield keyword to produce a series of results lazily, allowing us to iterate over values without storing them all in memory at once.

Here’s how we can implement this:

Powers of 2 Generator Function

def powers_of_two(exponent):  
    for i in range(exponent + 1):  # Loop from 0 to the given exponent  
        yield 2 ** i  # Yield 2 raised to the power of i  

# Example usage  
exponent = 10  # Specify the maximum exponent  
for power in powers_of_two(exponent):  
    print(power)  
Explanation of the Code
Generator Function Definition:

powers_of_two(exponent) is defined as a generator function that takes one parameter, exponent.
Looping Through Exponents:

The function uses a for loop that iterates from 0 to exponent (inclusive). This can be achieved by using range(exponent + 1).
Yielding Values:

Inside the loop, it computes 2 ** i (which is 2 raised to the power of i) and uses yield to return this value. The function will pause at yield and resume on the next call.
Example Usage
The example usage initializes a maximum exponent of 10 and iterates through the generator, printing each power of 2.

 7.Implement a generator function that reads a file line by line and yields each line as a string.


 Ans- We can create a generator function in Python that reads a file line by line and yields each line as a string. This approach is memory-efficient, as it allows us to process large files without loading the entire file into memory at once. Below is an implementation of such a generator function:

File Reading Generator Function

def read_file_line_by_line(file_path):  
    """A generator function that reads a file line by line."""  
    with open(file_path, 'r') as file:  # Open the file in read mode  
        for line in file:  # Iterate over each line in the file  
            yield line.strip()  # Yield the line after stripping whitespace  

# Example usage  
file_path = 'example.txt'  # Specify the path to our file  

try:  
    for line in read_file_line_by_line(file_path):  
        print(line)  # Print each line read from the file  
except FileNotFoundError:  
    print(f"The file '{file_path}' was not found.")  
Explanation of the Code
Generator Function Definition:

The function read_file_line_by_line(file_path) is defined to take one parameter, file_path, which should be the path to the file we want to read.
Opening the File:

The with open(file_path, 'r') as file: statement opens the file in read mode. Using with ensures that the file is properly closed after its suite finishes, even if an error occurs.
Reading Each Line:

The for line in file: loop goes through each line in the file.
The yield line.strip() statement ensures that each line is stripped of leading and trailing whitespace (including the newline character) before being returned.
Example Usage
The example usage specifies the path to a file (example.txt). It attempts to read the file line by line using the generator function and print each line. If the file is not found, a FileNotFoundError is caught, and an appropriate message is printed.

This generator function efficiently allows us to read long files line by line without overwhelming memory, as it only loads one line at a time into memory.




8.Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

Ans-We can use a lambda function in Python to sort a list of tuples based on the second element of each tuple by utilizing the sorted() function or the .sort() method of a list. Here’s how to do both:

Example

# Sample list of tuples  
data = [(1, 'apple'), (3, 'banana'), (2, 'orange'), (4, 'kiwi')]  

# Sorting using the sorted() function  
sorted_data = sorted(data, key=lambda x: x[1])  # Sort by the second element (fruit name)

data.sort(key=lambda x: x[1])  # Sort by the second element (fruit name)

Explanation of the Code
Sample Data:

We have a list of tuples named data, where each tuple consists of an integer and a string (fruit name).
Sorting using sorted():

The sorted() function returns a new sorted list from the elements of any iterable (in this case, data).
The key parameter specifies a function of one argument that is used to extract a comparison key from each list element. Here, we use a lambda function lambda x: x[1] which takes each tuple x and returns its second element.
Sorting using .sort():

The .sort() method sorts the list in place and modifies the original list. It also uses the same lambda function for the key parameter.
Output Sample

9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


Ans- We can use the map() function in Python to convert a list of temperatures from Celsius to Fahrenheit by applying a conversion formula to each temperature. The formula for converting Celsius to Fahrenheit is:

  Fahrenheit=(Celsius x 9/5)+32

Here’s how we can implement this in a Python program:

Program to Convert Celsius to Fahrenheit

# List of temperatures in Celsius  
celsius_temperatures = [0, 10, 20, 30, 40, 50]  

# Function to convert Celsius to Fahrenheit  
def celsius_to_fahrenheit(celsius):  
    return (celsius * 9/5) + 32  

# Using map() to apply the conversion function to each temperature  
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))  

# Print the converted temperatures  
print("Celsius temperatures:", celsius_temperatures)  
print("Fahrenheit temperatures:", fahrenheit_temperatures)  

Explanation of the Code

List of Celsius Temperatures:

The variable celsius_temperatures is a list that contains temperature values in Celsius.

Conversion Function:
The function .celsius_to_fahrenheit(celsius) takes a Celsius temperature as input and returns the corresponding Fahrenheit temperature using the formula.

Using map():
The map() function applies the conversion function to each element in the celsius_temperatures list. It returns a map object, which is converted to a list using list().
Output:

Finally, the program prints both the original list of Celsius temperatures and the converted list of Fahrenheit temperatures.

Output Sample
When we run the above program, the output will look like this:

Celsius temperatures: [0, 10, 20, 30, 40, 50]  
Fahrenheit temperatures: [32.0, 50.0, 68.0, 86.0, 104.0, 122.0]  

This demonstrates how the temperatures in Celsius were successfully converted to the corresponding Fahrenheit values using the map() function.




10.Create a Python program that uses `filter()` to remove all the vowels from a given string.


Ans- We can use the filter() function in Python to remove vowels from a given string by defining a function that checks whether each character is a consonant (i.e., not a vowel). Here's how we can implement this:

Program to Remove Vowels from a String


# Function to check if a character is a vowel  
def is_not_vowel(character):  
    return character.lower() not in 'aeiou'  

# Input string  
input_string = "Hello, World! This is a test string."  

# Using filter() to remove vowels  
filtered_characters = filter(is_not_vowel, input_string)  

# Joining the filtered characters back into a string  
result_string = ''.join(filtered_characters)  

# Print the result  
print("Original string:", input_string)  
print("String after removing vowels:", result_string)  

Explanation of the Code
Vowel Check Function:

The function is_not_vowel(character) checks if a character is a vowel. It converts the character to lowercase and checks if it is not in the string 'aeiou'.

Input String:

The variable input_string holds the string from which we want to remove the vowels.

Using filter():
The filter() function applies the is_not_vowel function to each character in input_string. It returns an iterator containing only the characters that return True (i.e., non-vowel characters).

Joining Characters:
The join() method is used to concatenate the filtered characters into a single string.

Output:

Finally, the program prints both the original string and the modified string with vowels removed.
Output Sample
When we run the above program, the output will look like this:

Original string: Hello, World! This is a test string.  
String after removing vowels: Hll, Wrld! Ths s t tst strng.  

This demonstrates how the program successfully removes all the vowels from the given input string.

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map



Ans-  This Programming calculates the total cost for each order and adjusts the amount based on the specified conditions, using a lambda function with map() to create a list of 2-tuples.

Python Program

# Sample data representing the orders  
# Each sublist contains: [Order Number, Book Title and Author, Quantity, Price per Item]  
orders = [  
    [34587, "Learning Python, Mark Lutz", 4, 40.95],  
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
